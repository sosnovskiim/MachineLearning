## Задача 1. Хитрый интеграл



### Условие

Какое наибольшее значение может принимать функционал
$$I(p) = \int\limits_0^{\infty}\exp\left( \frac{(x-4)^2 - (x+4)^2}{4} \right) \ln p(x) dx$$

при условии $p(x) > 0$, $\int\limits_0^{\infty} p(x) dx = 1$?

### Решение

Вычислим выражение в скобках:
$$
\frac{(x-4)^2 - (x+4)^2}{4} = \frac{(x^2 - 8x + 16) - (x^2 + 8x + 16)}{4} = \frac{-16x}{4} = -4x
$$

Следовательно,
$$
I(p) = \int_{0}^{\infty} e^{-4x} \ln p(x) \, dx
$$

Максимизируем $I(p)$ при ограничении $\int_{0}^{\infty} p(x) \, dx = 1$.

Функционал Лагранжа:
$$
L[p] = \int_{0}^{\infty} e^{-4x} \ln p(x) \, dx - \lambda \left( \int_{0}^{\infty} p(x) \, dx - 1 \right)
$$

Вариационная производная:
$$
\frac{\delta L}{\delta p(x)} = \frac{e^{-4x}}{p(x)} - \lambda = 0
$$

$$
p(x) = \frac{e^{-4x}}{\lambda}
$$

Из условия нормировки получаем:
$$
\int_{0}^{\infty} p(x) \, dx = \frac{1}{\lambda} \int_{0}^{\infty} e^{-4x} \, dx = 1
$$

$$
\int_{0}^{\infty} e^{-4x} \, dx = \frac{1}{4}
$$

$$
\frac{1}{\lambda} \cdot \frac{1}{4} = 1
$$

$$
\lambda = \frac{1}{4}
$$

Таким образом, оптимальная функция:
$$
p(x) = 4e^{-4x}
$$

Подставляем её в функционал:
$$
I(p) = \int_{0}^{\infty} e^{-4x} \ln\left(4e^{-4x}\right) dx = \int_{0}^{\infty} e^{-4x} \left(\ln 4 - 4x\right) dx = \ln 4 \int_{0}^{\infty} e^{-4x} dx - 4 \int_{0}^{\infty} x e^{-4x} dx
$$

$$
\int_{0}^{\infty} e^{-4x} dx = \frac{1}{4}, \quad
\int_{0}^{\infty} x e^{-4x} dx = \frac{1}{16}
$$

Подставляем:
$$
I(p) = \ln 4 \cdot \frac{1}{4} - 4 \cdot \frac{1}{16} = \frac{\ln 4}{4} - \frac{1}{4} = \frac{\ln 4 - 1}{4} \approx 0.097
$$

**Ответ** — 0.097

## Задача 2. Неоднозначный PCA

### Условие

Известно, что алгоритм PCA не всегда даёт однозначный ответ, потому что максимальный разброс может достигаться в проекции на несколько различных направляющих векторов.

Дан набор данных из 8 точек в трёхмерном пространстве, где точки являются вершинами 3-мерного куба. Для этого набора данных обучили PCA с одной главной компонентой (`n_components=1` в библиотеке sklearn). Какое количество различных ответов могло получиться? Иными словами, найдите количество направлений, проекции точек 3-мерного куба на которые имеют максимальный разброс. Направления, отличающиеся только знаком направляющего вектора, считаются различными.

### Решение

Есть 8 точек — вершины куба в $R^3$ с центром в начале координат и длиной ребра 2 (координаты $(\pm 1, \pm 1, \pm 1)$).   Нужно найти число направлений, дающих максимальную дисперсию проекции при PCA с `n_components = 1`. Направления, отличающиеся знаком, считаются разными.

Центр масс:  
$$
\frac{1}{8} \sum (\pm 1, \pm 1, \pm 1) = (0,0,0)
$$

Ковариационная матрица:
$$
\Sigma = \frac{1}{8} \sum_{i=1}^8 x_i x_i^T
$$

Диагональные элементы:  
$$
\Sigma_{11} = \frac{1}{8} \sum (\pm 1)^2 = 1 = \Sigma_{22} = \Sigma_{33}
$$

Недиагональные элементы:  
$$
\Sigma_{12} = \frac{1}{8} \sum (\pm 1 \cdot \pm 1) = 0 = \Sigma_{13} = \Sigma_{23}
$$

Таким образом, $\Sigma = E$ (единичная матрица 3x3).

Собственные значения: $\lambda_1 = \lambda_2 = \lambda_3 = 1$. Любой единичный вектор $v \in R^3$ — собственный.

Дисперсия проекции на любое направление:
$$
v^T \Sigma v = v^T E v = 1
$$

Как следствие, все направления дают одинаковую дисперсию проекции. PCA с одной компонентой может выбрать любое направление — все они являются главными компонентами.

Направление — единичный вектор в $R^3$. Направления, отличающиеся знаком, считаются разными. Множество единичных векторов в $R^3$ образует сферу $S^2$, имеющую бесконечное число точек.

**Ответ** — $\infty$.

## Задача 3. Метод ближайших соседей



### Условие

Рассмотрим $\ell$ точек, распределенных равномерно по объему $n$-мерного единичного шара с центром в нуле. Предположим, что мы хотим применить метод одного ближайшего соседа для точки $O$ — начала координат. Зададимся вопросом, на каком расстоянии будет расположен ближайший объект к точке $O$. Для ответа на этот вопрос выведите выражение для медианы расстояния от $O$ до ближайшего объекта.

Дайте ответ для $\ell = 500$ и $n=10$.

### Решение

Нужно найти медиану расстояния от начала координат $O$ до ближайшей из $\ell$ точек, равномерно распределённых в $n$-мерном единичном шаре.

Объём $n$-мерного шара радиуса $R$:
$$
V_n(R) = \frac{\pi^{n/2}}{\Gamma\left(\frac{n}{2}+1\right)} R^n
$$

А для $R=1$:
$$
V_n(1) = \frac{\pi^{n/2}}{\Gamma\left(\frac{n}{2}+1\right)}
$$

Вероятность того, что случайная точка лежит в шаре радиуса $r$:
$$
F(r) = \frac{V_n(r)}{V_n(1)} = r^n, \quad 0 \le r \le 1
$$

То есть $F(r) = r^n$ — функция распределения расстояния от центра до случайной точки.

Плотность распределения: $f(r) = n r^{n-1}, \quad 0 \le r \le 1$.

Пусть $R_1, R_2, \dots, R_\ell$ — расстояния до каждой из $\ell$ точек, независимые с функцией распределения $F(r) = r^n$.

Минимальное расстояние и функция распределения: $R_{\min} = \min(R_1, \dots, R_\ell)$.

$$
F_{\min}(r) = 1 - [1 - F(r)]^\ell = 1 - (1 - r^n)^\ell
$$

Посчитаем медиану $m$ $F_{\min}(m) = 0.5$:
$$
1 - (1 - m^n)^\ell = 0.5
$$

$$
(1 - m^n)^\ell = 0.5
$$

$$
1 - m^n = 0.5^{1/\ell}
$$

$$
m^n = 1 - 0.5^{1/\ell}
$$

$$
m = \left( 1 - 0.5^{1/\ell} \right)^{1/n}
$$

Теперь подставлим $\ell = 500$ и $n = 10$ и вычислим медиану $m$:
$$
0.5^{1/500} = e^{\frac{\ln 0.5}{500}}
$$

$$
e^{\frac{\ln 0.5}{500}} \approx 0.998614
$$

$$
m^{10} = 1 - 0.998615 = 0.001385
$$

$$
m = (0.001385)^{1/10} = \frac{\ln(0.001385)}{10} \approx -\frac{6.58206}{10} = -0.658206
$$

$$
e^{-0.658206} \approx 0.517779
$$

Медиана расстояния до ближайшего соседа от начала координат примерно равна $0.5$ в единицах радиуса шара.

**Ответ** — 0.5

## Задача 4. Градиентный спуск



### Условие

Пусть мы ищем минимум функции $f(x) = x^4$ с помощью градиентного спуска. Стартовая точка $x_0 = 4$. Используется обычный градиентный спуск, шаг которого выражается формулой $$x_{i+1} = x_i - \alpha\cdot\frac{d f}{d x}(x_i).$$ При каком минимальном значении $\alpha > 0$ алгоритм не сойдется?

Иными словами, при каком минимальном значении $\alpha > 0$ будет существовать такое $\delta > 0$, что алгоритм градиентного спуска, стартуя из точки $x_0$, никогда не придёт в интервал $(-\delta, \delta)$?

### Решение

Итерация градиентного спуска для $f(x) = x^4$ с начальной точкой $x_0 = 4$.
$$
x_{i+1} = x_i - \alpha f'(x_i), \quad f'(x) = 4x^3
$$

$$
x_{i+1} = x_i - \alpha \cdot 4x_i^3 = x_i (1 - 4\alpha x_i^2)
$$

Обозначим $g(x) = x (1 - 4\alpha x^2)$. Необходимо понять, при каких $\alpha$ траектория $x_{i+1} = g(x_i)$ не сходится к 0.

2-циклы: $x_2 = g(x_1), \quad x_1 = g(x_2)$, $x_1 \neq x_2 \neq 0$.  

$$
x_2 = x_1 (1 - 4\alpha x_1^2), \quad x_1 = x_2 (1 - 4\alpha x_2^2)
$$
  
$$
x_1 x_2 = x_1 x_2 (1 - 4\alpha x_1^2)(1 - 4\alpha x_2^2)
$$

При $x_1 x_2 \neq 0$:  
$$
1 = (1 - 4\alpha x_1^2)(1 - 4\alpha x_2^2)
$$

Обозначим $u = 4\alpha x_1^2$ и $v = 4\alpha x_2^2$. Тогда:
$$
(1-u)(1-v) = 1
$$

$$
uv = u + v
$$

$$
x_2 = x_1(1 - u)
$$

$$
v = u(1-u)^2
$$

$$
u \cdot u(1-u)^2 = u + u(1-u)^2
$$

$$
u^2 (1-u)^2 = u + u(1-u)^2
$$

Делим на $u \neq 0$:  
$$
u (1-u)^2 = 1 + (1-u)^2
$$

$$
u (1 - 2u + u^2) = 1 + 1 - 2u + u^2
$$

$$
u - 2u^2 + u^3 = 2 - 2u + u^2
$$

$$
u^3 - 3u^2 + 3u - 2 = 0
$$

Проверяем для $u=2$:
$$
8 - 12 + 6 - 2 = 0
$$

Делим на $u-2$:
$$
u^2 - u + 1 = 0
$$
Получаем, что дискриминант отрицателен. Т. е. $u = 2$ — единственное вещественное решение. Тогда:
$$
u = 4\alpha x_1^2 = 2
$$

$$
x_1^2 = \frac{1}{2\alpha}
$$

$$
v = u(1-u)^2 = 2(1-2)^2 = 2
$$

$$
x_2^2 = \frac{1}{2\alpha}
$$

Рассмотрим $|g(x)| = |x| \cdot |1 - 4\alpha x^2|$.

Если $|1 - 4\alpha x^2| > 1$, то $|x_{i+1}| > |x_i|$.

Это неравенство $|1 - 4\alpha x^2| > 1$ выполняется, когда $1 - 4\alpha x^2 < -1$:

$$
1 - 4\alpha x^2 < -1
$$

$$
4\alpha x^2 > 2
$$

$$
x^2 > \frac{1}{2\alpha}
$$

То есть если $|x_i| > \frac{1}{\sqrt{2\alpha}}$, то $|x_{i+1}| > |x_i|$.

Применяем к $x_0 = 4$.

Если $4 > \frac{1}{\sqrt{2\alpha}}$, то $|x_1| > |x_0|$, и процесс уходит от 0.

Условие $4 > \frac{1}{\sqrt{2\alpha}}$ равносильно условию:  
$$
\frac{1}{\sqrt{2\alpha}} < 4
$$

$$
\sqrt{2\alpha} > \frac14
$$

$$
2\alpha > \frac{1}{16}
$$

$$
\alpha > \frac{1}{32}
$$

При $\alpha = \frac{1}{32}$: $\frac{1}{\sqrt{2\alpha}} = \frac{1}{\sqrt{1/16}} = 4$:  
$$
x_1 = 4(1 - 4\cdot \frac{1}{32} \cdot 16) = 4(1 - 2) = -4
$$

$$
x_2 = -4(1 - 4\cdot \frac{1}{32} \cdot 64) = -4(1 - 8) = 28
$$
То есть для $x_2$ получаем расходимость.

**Ответ** — минимальное $\alpha > 0$, при котором алгоритм не сходится к 0:
$$
\alpha = \frac{1}{32}
$$

## Задача 6. Мандаринки

### Условие


Тася и ее друг Ваня очень любят мандарины, но Тася предпочитает абхазские, а Ваня — марокканские. Многочисленные эксперименты с мандаринами позволили ребятам вывести четыре основных показателя внешнего вида мандарина:


* толщина мандарина $x_1$ (численный признак);
* диаметр наибольшего сечения мандарина $x_2$ (численный признак);
* насыщенность цвета $x_3$ (численный признак);
* наличие/отсутствие листика $x_4$ (бинарный признак, принимает значения 0/1).

Численные признаки приведены в особых единицах, поэтому могут принимать как положительные, так и отрицательные значения. Более того, ребята установили, что распределение вектора параметров $(x_1, x_2, x_3, x_4)$ следующее: для обоих сортов вектор $(x_1, x_2, x_3)$ имеет трехмерное нормальное распределение, причём
* для абхазских манадринов — с вектором средних $a_1 = (0,0,0)$ и матрицей ковариаций
$K_1=\begin{pmatrix} 1 & 1 & 0\\1 & 2 & 1\\0 & 1 & 2\end{pmatrix}$;

* для марокканских — с вектором средних $a_2 = (6,6,6)$ и матрицей ковариаций
$K_2 = \begin{pmatrix} 1 & 0 & 0\\0 & 1 & 0\\0 & 0 & 1\end{pmatrix}$.

Наконец, признак $x_4$ не зависит от совокупности признаков $(x_1, x_2, x_3)$ для обоих сортов, причём для абхазского имеет распределение $\mathrm{Bern}(p_1)$, а для марокканского — $\mathrm{Bern}(p_2)$ (здесь $p_i$ — вероятность наличия листика у мандаринки $i$-ого класса, $i=1,2$), $p_1 = 1/2$, $p_2 = 1/3$.

Тася купила мандарины двух видов, оставила сумки на столе и пошла отдохнуть, но, пока она спала, кошка Мурка взобралась на стол, вывалила мандарины из обеих сумок и перемешала их. Помогите ребятам построить дискриминативную функцию методом максимального правдоподобия!





### Решение

Пусть $y = 1$ — абхазские мандарины, а $y=2$ — марокканские.

При равных априорных вероятностях $P(y = 1) = P(y = 2) получаем бинарную классификацию:
$$
\frac{p(x \mid y = 1)}{p(x \mid y = 2)} > 1
$$

Логарифмируем и получаем, что:
$$
\ln p(x \mid y = 1) - \ln p(x \mid y = 2) > 0
$$

Параметры распределения для $(x_1, x_2, x_3)$:
- $y=1$: $N(a_1, K_1)$, где $a_1 = (0,0,0)$,
  $K_1 = \begin{pmatrix} 1 & 1 & 0 \\ 1 & 2 & 1 \\ 0 & 1 & 2 \end{pmatrix}$

- $y=2$: $N(a_2, K_2)$, где $a_2 = (6,6,6)$, 
  $K_2 = E_3$

Для $x_4$ независимого от $(x_1,x_2,x_3)$:
- $y=1$: $Bern(p_1)$, $p_1 = \frac12$

- $y=2$: $Bern(p_2)$, $p_2 = \frac13$

При равных априорных вероятностях получаем следующий логарифм отношения правдоподобия:
$$
L(x) = \ln p(x \mid y=1) - \ln p(x \mid y=2) = \ln p(x_1,x_2,x_3 \mid y=1) - \ln p(x_1,x_2,x_3 \mid y=2) + \ln\frac{p(x_4 \mid y=1)}{p(x_4 \mid y=2)}
$$

Получим гауссову часть для многомерного нормального распределения:
$$
\ln p(x_1,x_2,x_3 \mid y=i) = -\frac{3}{2}\ln(2\pi) - \frac12 \ln|K_i| - \frac12 (x-a_i)^T K_i^{-1} (x-a_i)
$$

Разность гауссовых частей:
$$
G(x) = -\frac12 \ln|K_1| + \frac12 \ln|K_2| - \frac12 (x-a_1)^T K_1^{-1} (x-a_1) + \frac12 (x-a_2)^T K_2^{-1} (x-a_2)
$$

Посчитаем определители матриц:
$$
|K_1| = 1, \quad |K_2| = 1
$$

$$
-\frac12 \ln|K_1| + \frac12 \ln|K_2| = 0
$$

Также получим обратную матрицу $K_1^{-1}$:
$$
K_1^{-1} = \begin{pmatrix} 3 & -2 & 1 \\ -2 & 2 & -1 \\ 1 & -1 & 1 \end{pmatrix}
$$

Найдём квадратичные формы. Обозначим:
$$
u = (x_1, x_2, x_3)
$$

$$
v = (x_1-6, x_2-6, x_3-6)
$$

Тогда:
$$
Q_1 = u^T K_1^{-1} u = 3x_1^2 - 4x_1x_2 + 2x_1x_3 + 2x_2^2 - 2x_2x_3 + x_3^2
$$

$$
Q_2 = v^T v = (x_1-6)^2 + (x_2-6)^2 + (x_3-6)^2
$$

$$
G(x) = -\frac12 Q_1 + \frac12 Q_2
$$

Бинарная часть:
$$
B(x_4) = \ln\frac{p(x_4 \mid y=1)}{p(x_4 \mid y=2)}
$$

Если $x_4 = 1$:
$$
B(1) = \ln\frac{1/2}{1/3} = \ln\frac{3}{2}
$$

Если $x_4 = 0$:
$$
B(0) = \ln\frac{1/2}{2/3} = \ln\frac{3}{4}
$$

Выпишем итоговую дискриминантную функцию:
$$
L(x) = G(x) + B(x_4) = -\frac12 Q_1 + \frac12 Q_2 + B(x_4)
$$
