### Упражнение 4.1

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def corr_cos(x, y):
    x = np.array(x, dtype=float)
    y = np.array(y, dtype=float)
    
    cos_sim = np.sum(x * y) / (np.sqrt(np.sum(x ** 2)) * np.sqrt(np.sum(y ** 2)))
    
    x_mean = np.mean(x)
    y_mean = np.mean(y)
    
    x_centered = x - x_mean
    y_centered = y - y_mean
    
    pearson_corr = np.sum(x_centered * y_centered) / (
        np.sqrt(np.sum(x_centered ** 2)) * np.sqrt(np.sum(y_centered ** 2))
    )
    
    return pearson_corr, cos_sim
```

Центрированные векторы

```python
x1 = np.array([-2, -1, 0, 1, 2])
y1 = np.array([-4, -2, 0, 2, 4])
pearson_1, cos_sim_1 = corr_cos(x1, y1)
print(f"Вектор x: {x1}")
print(f"Вектор y: {y1}")
print(f"Корреляция Пирсона: {pearson_1:.6f}")
print(f"Косинусное сходство: {cos_sim_1:.6f}")
```

Неценрированные векторы

```python
x2 = np.array([1, 2, 3, 4, 5])
y2 = np.array([1, 3, 2, 5, 4])
pearson_2, cos_sim_2 = corr_cos(x2, y2)
print(f"Вектор x: {x2}")
print(f"Вектор y: {y2}")
print(f"Корреляция Пирсона: {pearson_2:.6f}")
print(f"Косинусное сходство: {cos_sim_2:.6f}")
```

### Упражнение 4.2

```python
base_vector = np.array([0, 1, 2, 3])
shifts = np.arange(-50, 51)
pearson_array = []
cos_sim_array = []
for shift in shifts:
    pearson, cos_sim = corr_cos(base_vector, base_vector + shift)
    pearson_array.append(pearson)
    cos_sim_array.append(cos_sim)

plt.figure(figsize=(8, 6))
plt.plot(shifts, pearson_array, label='Корреляция Пирсона')
plt.plot(shifts, cos_sim_array, label='Косинусное сходство')
plt.xlabel('Смещение')
plt.ylabel('Значение')
plt.title('Влияние смещения на корреляцию и косинусное сходство')
plt.legend()
plt.grid(True, alpha=0.3);
```
