# Вариант 2
Заполнить матрицу случайными числами.

```python
def generate_matrix_list(n):
    """
    Функция генерирует и возвращает двумерную матрицу
    размерности n x n, состоящую из псевдослучайных трёхзначных
    чисел, с помощью списочных выражений. Принимает один
    позиционный аргумент - размерность генерируемой матрицы.
    """
    return [[randint(100, 999) for _ in range(n)] for _ in range(n)]


def generate_matrix_array(n):
    """
    Функция генерирует и возвращает двумерную матрицу
    размерности n x n, состоящую из псевдослучайных трёхзначных
    чисел, с помощью векторизованных функций numpy. Принимает один
    позиционный аргумент - размерность генерируемой матрицы.
    """
    return np.random.randint(100, 1000, (n, n))
```

Генерация матрицы 10x10

```python
"""
Замеряется среднее время выполнения
генерации матрицы размерности 10х10
с помощью списочных выражений.
"""
%timeit generate_matrix_list(10)
```

```python
"""
Замеряется среднее время выполнения
генерации матрицы размерности 10х10
с помощью векторизованных функций numpy.
"""
%timeit generate_matrix_array(10)
```

Генерация матрицы 100x100

```python
"""
Замеряется среднее время выполнения
генерации матрицы размерности 100х100
с помощью списочных выражений.
"""
%timeit generate_matrix_list(100)
```

```python
"""
Замеряется среднее время выполнения
генерации матрицы размерности 100х100
с помощью векторизованных функций numpy.
"""
%timeit generate_matrix_array(100)
```
