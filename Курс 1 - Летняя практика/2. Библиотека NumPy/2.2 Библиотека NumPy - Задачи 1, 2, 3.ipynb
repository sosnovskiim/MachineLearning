## 9. Заданный одномерный массив
Дан одномерный массив числовых значений, насчитывающий N элементов. Определить образуют ли элементы массива, расположенные перед первым отрицательным элементом, возрастающую последовательность

```python
""" Задаётся одномерный массив, состоящий из целых чисел. """
r = np.array([1, 2, 3, 4, -5, 3])

""" Находится индекс первого отрицательного числа. """
f = np.nonzero(r < 0)[0][0]

"""
Определяется, образуют ли возрастающую последовательность числа,
расположенные до первого отрицательного числа.

С помощью функции diff получается массив разниц значений
между соседними элементами. Если набор чисел, образованный
функцией sign из этого массива, состоит только из единиц,
то последовательность возрастающая.
"""
print(np.all(np.sign(np.diff(r[:f])) == 1))
```

## 10. Заданная матрица
Дана матрица. Выяснить является ли сумма элементов, расположенных над нулями, чётным числом.

```python
""" Задаётся матрица, состоящая из целых чисел. """
m = np.array([[4, 0, 4, 2],
              [0, 9, 0, 5],
              [8, 0, 7, 0]])

""" Транспонируется ранее заданная матрица. """
m = m.T

""" Получаются массивы с индексами нулевых элементов. """
z = np.nonzero(m == 0)
x, y = z[1], z[0]

"""
Все индексы по строкам уменьшаются на единицу.
Если индекс равен 0, то остаётся без изменений.
"""
x = x - (x != 0)

"""
Суммируются все числа, находящиеся слева от нулей,
и их сумма проверяется на чётность.
"""
print(np.sum(m[y, x]) % 2 == 0)
```
