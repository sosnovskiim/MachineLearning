# Вариант 2
Дан вектор x размерности N и матрица A размерности 2 * N x N. Умножить первую половину строк матрицы A на вектор x. Каждый столбец из второй половины строк матрицы A умножить на вектор x. Результат записать в две новые матрицы и посчитать их произведение.

```python
"""
Задаётся число N и генерируются вектор x размерности N
и матрица a размерности 2 * N x N, состоящие из целых чисел.
"""
n = 4
x = np.random.randint(-10, 10, n)
a = np.random.randint(-10, 10, (2 * n, n))

"""
Считаются произведение строк первой половины матрицы a на вектор x
и произведение столбцов второй половины матрицы a на вектор x.
"""
m1 = np.dot(a[:n], x)
m2 = np.dot(a[n:].T, x)

""" Выводится произведение ранее посчитанных матриц. """
print(np.dot(m1, m2))
```
